```python
columns = ['is_canceled', 'arrival_date_year', 'arrival_date_week_number', 'arrival_date_day_of_month', 'is_repeated_guest',
          'agent', 'company']

for i in columns:
    df[i] = df[i].astype('category')

df.dtypes
```

```python
df.isnull().sum()
```

```python
columns = ['children', 'country_origin', 'agent', 'company']

for i in columns:
    try:
        df[i].skew()
        if df[i].skew() == 0:
            df[i] = df[i].fillna(df[i].mean())
        elif df[i].skew() != 0:
            df[i] = df[i].fillna(df[i].median())
    except TypeError:
        df[i] = df[i].fillna(df[i].mode()[0])

df.isnull().sum()
```

```python
df_nocancel = df[df['is_canceled'] == 0]
df_nocancel.shape
```

```python
df_country_count = df_nocancel[df_nocancel['is_repeated_guest']==0]['country_origin'].value_counts()
df_country_count
```

```python
df_country = pd.DataFrame(df_country_count).reset_index()
df_country.rename(columns={'index':'code', 'country_origin':'total_guest'}, inplace=True)
df_country
```

```python
alpha2 = [i.alpha_2 for i in list(pycountry.countries)]
alpha3 = [i.alpha_3 for i in list(pycountry.countries)]
```

```python
def country_code(df):
    if (len(df['code'])==2 and df['code'] in alpha2):
        return pycountry.countries.get(alpha_2= df['code']).name
    elif (len(df['code'])==3 and df['code'] in alpha3):
        return pycountry.countries.get(alpha_3= df['code']).name
    else:
        return 'Code is not found'

df_country['country_name'] = df_country.apply(country_code, axis=1)
df_country
```

```python
df_guest = df_nocancel[(df_nocancel.loc[:,'adults'] > 0) & (df_nocancel.loc[:,'children'] >= 0)]
df_guest.shape
```

```python
df_guest.loc[:,'total_guest'] = df_guest.loc[:,'adults'] + df_guest.loc[:,'children']
df_guest.loc[:,'guest_pay'] = df_guest.loc[:,'adr'] / df_guest.loc[:,'total_guest']
df_guest.shape
```

```python
mean_room = df_guest.groupby('reserved_room_type')['guest_pay'].mean()
df_room = pd.DataFrame(mean_room).sort_values('guest_pay').reset_index()
df_room
```

```python
df_month = df_guest.loc[:, ['arrival_date_month', 'guest_pay']]
month = ['January', 'February', 'March', 'April', 'May', 'June',
        'July', 'August', 'September', 'October', 'November', 'December']
df_month.loc[:,'arrival_date_month'] = pd.Categorical(df_month.loc[:,'arrival_date_month'], categories = month)

plt.figure(figsize = (15,8))
sns.lineplot(x = 'arrival_date_month', y = 'guest_pay', data = df_month)
plt.title('Room price per night over the year', fontsize = 20)
plt.xlabel('Month', fontsize = 15)
plt.ylabel('Price', fontsize = 15)
```

```python
guest_month = df_nocancel['arrival_date_month'].value_counts()

df_guest_month = pd.DataFrame(guest_month).reset_index()
df_guest_month.rename(columns={'index':'month', 'arrival_date_month':'total_guest'}, inplace=True)

month = ['January', 'February', 'March', 'April', 'May', 'June',
        'July', 'August', 'September', 'October', 'November', 'December']
df_guest_month.loc[:,'month'] = pd.Categorical(df_guest_month.loc[:,'month'], categories = month)

plt.figure(figsize = (15,8))
sns.barplot(x = 'month', y = 'total_guest', data = df_guest_month)
plt.title('Total Guests per Month', fontsize = 20)
plt.xlabel('Month', fontsize = 15)
plt.ylabel('Guests', fontsize = 15)
```

```python
cancellation = df['is_canceled'].value_counts()
df_cancellation = pd.DataFrame(cancellation).reset_index()
df_cancellation.rename(columns={'index':'cancellation', 'is_canceled':'frequency'}, inplace=True)
df_cancellation['cancellation'] = df_cancellation['cancellation'].astype('object')
df_cancellation.loc[(df_cancellation['cancellation'] == 0), 'cancellation'] = 'No'
df_cancellation.loc[(df_cancellation['cancellation'] == 1), 'cancellation'] = 'Yes'
df_cancellation
```

```python
df_cancellation['percent'] = [round((df_cancellation.frequency[0] / df_cancellation.frequency.sum())*100, 2), 
                              round((df_cancellation.frequency[1] / df_cancellation.frequency.sum())*100, 2)]
df_cancellation
```
